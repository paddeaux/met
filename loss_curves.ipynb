```python
import pandas as pd
import matplotlib.pyplot as plt
import csv

gen_loss = pd.read_csv('gen_loss.csv')

print(gen_loss)
gen_loss['batch_number'] = gen_loss['batch_number']/2
gen_64 = gen_loss[gen_loss['size']==64.0]

critic_loss = pd.read_csv('critic_loss.csv')
critic_loss['batch_number'] = critic_loss['batch_number']/2
critic_64 = critic_loss[critic_loss['size']==64.0]

fig, ax = plt.subplots(1,1,figsize=(10,8))

critic_64.plot.scatter(ax=ax, x='epoch', y='loss', c='orange',s=0.1)
gen_64.plot.scatter(ax=ax, x='epoch', y='loss', c='blue',s=0.1)

plt.show()

```
