```python
import pandas as pd
import matplotlib.pyplot as plt
import csv

gen_loss = pd.read_csv('generator_256_loss.csv')
critic_loss = pd.read_csv('critic_256_loss.csv')
```

```python
fig, ax = plt.subplots(1,1,figsize=(15,5))

critic_loss[critic_loss['epoch'] ==98].plot(ax=ax, x='batch_number', y='loss', c='orange', label='Critic Loss')
gen_loss[gen_loss['epoch'] == 98].plot(ax=ax, x='batch_number', y='loss', c='blue', label='Generator Loss')
ax.legend()
plt.show()
```
