```python
import pandas as pd
import matplotlib.pyplot as plt
import csv

gen_loss = pd.read_csv('gen_loss.csv')
gen_loss['batch_number'] = gen_loss['batch_number']/2
gen_256 = gen_loss[gen_loss['size']==256.0]

critic_loss = pd.read_csv('critic_loss.csv')
critic_loss['batch_number'] = critic_loss['batch_number']/2
critic_256 = critic_loss[critic_loss['size']==256.0]

fig, ax = plt.subplots(1,1,figsize=(10,8))

critic_256[critic_256['loss'] < 500].plot.scatter(ax=ax, x='batch_number', y='loss', c='orange',s=0.1)
gen_256[gen_256['loss'] < 500].plot.scatter(ax=ax, x='batch_number', y='loss', c='blue',s=0.1)

plt.show()

```
